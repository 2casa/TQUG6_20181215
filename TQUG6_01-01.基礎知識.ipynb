# 用語の説明

まずは、今日の勉強会で使われる養母の説明を簡単に行いたいと思います。

### ファンダメンタル

日本語では「基本的な」という意味。経済、特に投資の世界においては分析対象企業の財務情報や、経営情報など、当該企業の「基礎的な情報」のことを「ファンダメンタル」と呼びます。

そうした基礎的情報から、企業「価値」を計算し、投資判断を行うことをファンダメンタル分析と呼びます。

取引所で売買されている株式「価格」が企業「価値」よりも低ければ「割安」。
逆であれば高ければ「割高」と判断できます。

## 「価値」と「価格」の違い

「価値」と「価格」は厳密には違います。一言でその違いを説明するならば、

* 「価値(Value)」は得られるもの。
* 「価格(Price)」は支払うもの。

ということができると思います。

ところが、一般的には価値 = 価格 には「ならない」ことがあります。どういう時でしょうか？

### テクニカル

Value = Price に「ならない」理由の一つが、「需要と供給」の不均衡。
需要>供給であれば、 「多少割高であっても買う」という人が現れるため、価格が（価値の変化とは無関係に）上昇する。


### ファクター

日本語では「要因」という意味。今回の勉強会において「ファクター」と言えば、株式の収益（リターン）に影響を与えるような「要因」のことをファクターと呼びます。


### ファクターとアノマリーは何が違うのか？

「アノマリー」から「ファクター」へと昇格するためには、以下の基準を満たさないといけないとされる。

（ファクター投資入門より）

* 持続性
   * 長期間にわたり、異なる経済的レジームでも有効であること
* 普遍性
  * あらゆる国、地域、セクター、さらにはアセットクラスで有効である
* 安定性
  * どのような定義でも有効である
* 投資可能性
  * 机上のみならず、取引コストなど実践するときの検討事項を考慮した後でも有効である
* 合理的な説明
  * そのプレミアムとそれが存続する理由を、リスクに基づき、または投資家の行動に基づいて、合理的に説明することができる

### サイズファクター
「小型株効果」という言葉を知っているもいるかと思います。

ファクター：ロングショートポートフォリオで構築される。
サイズファクター：小型株をロング / 大型株をショート

* 財務体制が脆弱であるため、金融引き締め / 緩和時　の波がある
  * サイズファクターのプレミアムは、経済サイクルのリスクを取ることによって得ていると解釈できる。
    * 「市場インデックス以上にリスクを取らなければならないのであれば、投資家はそれ以上の期待リターンを要求する」という風に解釈します。

### グロース株
「バリュー株効果」の逆

* バリュー株：低PER
* グロース株：高PER

$$ PER =  \frac{株価}{一株あたり利益} $$ 

* 低PER : 利益に対して株価が（相対的に）安いと考えられるので、バリュー（割安）株と呼ばれる
* 高PER : 利益に対して株価が（相対的に)高い。投資家は、現在の利益で評価せず将来性を評価していると考えられるので、グロース（成長）株と呼ばれる

こういう計算が、Quantopianでは簡単にできます。Part.2 でやる予定。
